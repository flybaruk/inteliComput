```python
import pandas as pd
import io
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('simulacao-Main.csv')

# --- Passo 2: Calcular a m√©dia de 'Score' a cada 10 runs ---
df['run_group'] = (df.index // 10)
average_scores = df.groupby('run_group')['Score'].mean()


# --- Passo 3: Preparar os dados para o gr√°fico ---
# Criar uma lista de r√≥tulos descritivos para o eixo X do gr√°fico
group_labels = [f'Runs {i*10+1}-{(i+1)*10}' for i in average_scores.index]


# --- Passo 4: Criar o gr√°fico de barras dos grupos ---
plt.figure(figsize=(10, 6))

# Usamos os r√≥tulos que criamos no eixo X e as m√©dias no eixo Y
plt.bar(group_labels, average_scores, color='teal')

# Adiciona uma linha horizontal no zero
plt.axhline(0, color='grey', linewidth=0.8)

# Adiciona t√≠tulos e r√≥tulos
plt.title('M√©dia de Score por Grupo de 10 Simula√ß√µes Main')
plt.xlabel('Grupo de Simula√ß√µes (Runs)')
plt.ylabel('M√©dia de Score')

# Rotaciona os r√≥tulos do eixo X para evitar sobreposi√ß√£o
plt.xticks(rotation=45, ha='right')

# Ajusta o layout para que os r√≥tulos n√£o sejam cortados
plt.tight_layout()


# --- Passo 5: Salvar o gr√°fico ---
plt.savefig('grafico_barras_grupos_Main.png')

# Se desejar exibir o gr√°fico na tela ao rodar localmente:
# plt.show()

print("Gr√°fico de barras dos grupos gerado e salvo como 'grafico_barras_grupos_Main.png'")
```
